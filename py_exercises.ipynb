# 3. bubble sort

Sort a list in ths mentioned algorithm -- don't look in internet


image.png

```python
list1=[5,1,4,2,8]
def bubbleSort(list1):
  n=len(list1)
  for i in range(0,n):
    for j in range(0,n-1):
      if list1[j]>list1[j+1]:
        list1[j],list1[j+1]=list1[j+1],list1[j]
print(*list1, sep=",")

bubbleSort(list1)
print("the sorted list is")
for i in range(len(list1)):
  print(list1[i], end=',')
```

# 5. encrypt and decrypt a string

A string is considered to be a sequence of integers with ASCII value equivalent of each characters. A moving sum of length 3 is done to encrypt the string and the resulting list of integers is decrypted back to the original input

Example:
* inptxt = "office space" 
* is ecrypted to
* encryp = [315, 309, 306, 305, 232, 248, 259, 324, 308, 297, 200, 101]
* and decrypted back to 
* outtxt = "office space"



```python
#Step1: Analyse the data and the problem
#Step2: Mentally execute the problem/on paper
#Step3: Step by Step execution

#inptxt="o"
#print("The ASCII value is", ord(inptxt))

inptxt="of"
inp_txt=list(inptxt)
print(inp_txt)
av=[]
enc = []
for i in inp_txt:
  av.append(ord(i))
  #print(av)
  if len(av) >= 3:
    sum = av[-1] + av[-2] + av[-3]
    enc.append(sum)
if len(av) > 1:
  enc = enc + [av[-1] + av[-2], av[-1]]
else:
  enc = av
print(enc)
outtxt=''
output_list = []

rev_enc = enc[::-1]
#print(rev_enc)

for l in range(len(rev_enc)):
  if l == 0 :
    output_list.append(rev_enc[l])
  if l == 1 or l ==2 :
    output_list.append(rev_enc[l] - rev_enc[l-1])
  if l > 2:
    output_list.append(rev_enc[l] - (output_list[-1] + output_list[-2]))
  #print(output_list) 
output_list = output_list[::-1]
for k in output_list:
  outtxt = outtxt + chr(k)
print(outtxt)


  
```
